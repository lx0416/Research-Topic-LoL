## Research Question: Which champion has the highest attack damage (attack speed inclusive)? 

### Combine and create a new column 'totalattack'
- Calculated using 4 columns of data in the dataset
    - attackdamage, attackdamageperlevel, attackspeed, attackspeedperlevel
    - Background information: The highest level a champion can reach is 18 in a gameplay
    - Formula: (attackdamage + (attackdamageperlevel x 17)) x (attackspeed x (attackspeedperlevel x 17 / 100) + 1))

```python
chdata['totalattack'] = (chdata['attackdamage'] + 17*chdata['attackdamageperlevel']) * (((chdata['attackspeedperlevel']*17/100)+1) * chdata['attackspeed'])
```

### Shows the newly created column 'totalattack'

```python
chdata
```

### Finding the champion with the highest attack in consideration with the attack speed

```python
chdata[chdata['totalattack'] == chdata['totalattack'].max()]
```

As can be seen, the champion with the highest attack damage (considering attack speed) is Malphite

## Data Pipeline
- Merged different columns of data for the various dataframes created that relate to the research question of which champions have higher damage (including attack speed) stated above
    - Merged and calculated the total attack by integrating and processing the 4 aspects of total damage
- The  processed data on totalattack is visually represented in a barchart below

### Graphically Represented totalattack
- A new dataframe is created and assigned for efficient sorting
- Sorted in descending order, according to totalattack
- When represented in a barchart, it is easier to analyse and determine which are the top few champions with the highest attack damage

```python
chdesc = chdata.sort_values('totalattack',ascending=True)

#plotting barchart
fig = plt.figure()
ax = fig.add_axes([0,0,2,8])
name = chdesc['id']
attack = chdesc['totalattack']
y_pos = np.arange(len(name))

plt.barh(y_pos, attack, align='center', alpha=0.5, color='red')
plt.yticks(y_pos, name)
plt.xlabel('Total Attack')
plt.ylabel('Champion')
plt.title('Champions sorted by totalattack')

# Function to round up to 2 decimal places 
def truncate(n, decimals=0):
    multiplier = 10 ** decimals
    return int(n * multiplier) / multiplier

# Printing of rounded up totalattack values
for i, v in enumerate(attack):
    ax.text(v + 0.5, i - 0.2, str(truncate(v, 2)), color='black', fontweight='bold')

plt.show()
```

### Interesting Insight/Finding
- Jhin is a marksman, which means he is one of many ADCs (attack damage carry), that carries his team and the game due to his high damage
- Highest pmdamage (physical/magic damage) belongs to champion Jhin, however, the one with the highest attack is Malphite. Why?
    - Jhin has 0 attackspeedperlevel, while Malphite has an attackspeedperlevel of 2.67
    - This means that Jhin does not have high pmdamage due to his attack speed but only due to his physical damage
    - Thus, we can infer from this that to build Jhin well, his equipment will require high damage, critical chance and critical damage, and not attackspeed

## Research Question: Which champion is the highest magic damage dealer?

### First, compile all champions with the related data columns
- It is appropriate to create a new dataframe here so that we do not obfuscate previous steps in our pipeline
- We require the name, tags and totalattack columns

```python
#creating a copy of the original dataset with select columns
ntt_df = chdata.loc[:,('id', 'tags', 'totalattack')]

#creating a new column called mage and assigning 0 to all rows
ntt_df['mage'] = '0'

ntt_df
```

### Secondly, we locate the champions with the Mage tags
- We create a mage data column which indicates if the champion has the mage tag or not
- 1 represents that the champion is a mage, while 0 indicates the champion is not a mage

```python
#replacing the values in mage column to 1 if they contain the mage tag
ntt_df["mage"] = ntt_df["tags"].map(lambda x: "1" if "Mage" in x else "0" )
ntt_df
```

### Thirdly, we compress the data to a sub dataset with only mages
- We compile only the champions with a value of 1 in the mage data column

```python
magedf = ntt_df[ntt_df['mage'].isin(['1'])]
magedf
```

### Number of mages (out of a total of 149 champions)

```python
len(magedf)
```

### Graphically Represented totalattack for Mages
- A new dataframe is created and assigned for efficient sorting
- For quick and easy understanding, we visually represent the results in a barchart
- Sorted in descending order, according to totalattack

```python
magedf_desc= magedf.sort_values('totalattack',ascending=True)
```

```python
#plotting barchart
fig = plt.figure()
ax = fig.add_axes([0,0,2,4])
objects = magedf_desc['id']
y_pos = np.arange(len(objects))
attack = magedf_desc['totalattack']

plt.barh(y_pos, attack, align='center', alpha=0.5, color='purple')
plt.yticks(y_pos, objects)
plt.xlabel('Total Attack')
plt.ylabel('Champion')
plt.title('Mages sorted by totalattack')

for i, v in enumerate(attack):
    ax.text(v + 0.5, i , str(truncate(v, 2)), color='black', fontweight='bold')

plt.show()
```

As can be seen above in the barchart, the title of highest magic damage dealer belongs to Kassadin
